# Dilatação

```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
import numpy as np


uploaded = files.upload()
img = cv2.imread('imagem4.png')
kernel = np.ones((5, 5), np.uint8)

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(tresh, img) = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)


cv2_imshow(img)
img_dilation = cv2.dilate(img, kernel, iterations=3)
cv2_imshow(img_dilation)
```

# Erosão

```python
uploaded = files.upload()
img = cv2.imread('imagem1.png')

img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(tresh, img_pb) = cv2.threshold(img_cinza, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((3, 3), np.uint8)

img_erode = cv2.dilate(img_pb, kernel, iterations=2)
cv2_imshow(img)
cv2_imshow(img_cinza)
cv2_imshow(img_pb)
cv2_imshow(img_erode)

# Abertura e fechamento
img_abertura = cv2.morphologyEx(img_erode, cv2.MORPH_OPEN, kernel)
img_fechamento = cv2.morphologyEx(img_abertura, cv2.MORPH_CLOSE, kernel)
cv2_imshow(img_abertura) # Melhor alternativa
cv2_imshow(img_fechamento)
```
